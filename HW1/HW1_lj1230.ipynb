```python
import csv

def csvRows(filename):
    with open(filename, 'r') as fi:
        reader = csv.DictReader(fi)
        for row in reader:
            yield row["Product ID"], row["Customer ID"], float(row["Item Cost"])

ProductDict = {}
for ProductID, CustomerID, Revenue in csvRows('sale.csv'):
    if ProductID not in ProductDict:
        ProductDict[ProductID] = [[CustomerID], Revenue]
    else:
        if CustomerID not in ProductDict[ProductID][0]:
            ProductDict[ProductID][0].append(CustomerID)
        ProductDict[ProductID][1] += Revenue

ProductDict
```

```python
# 因为是按照C ID排下来的，所以就可以省略查找这一步
# 空间节约了很多，因为无需额外存储排好序的c id
import csv

lastID = None
currentSet = None
customers = {}
revenues = {}
with open('sale.csv', 'r') as fi:
    reader = csv.DictReader(fi)
    for d in reader:
        if lastID != d["Customer ID"]:
            if lastID != None:
                for p in currentSet:
                    customers[p] = customers.get(p, 0) + 1  # 换了一个新的c id时，把上一个c id对应的p id做一个统计
            lastID = d["Customer ID"]
            currentSet = set()  # 换了一个c id时就清0
        currentSet.add(d["Product ID"])  # 计算了每个c id对应有多少distinct的p id
        revenues[d["Product ID"]] = revenues.get(d["Product ID"], 0) + float(d["Item Cost"])
if lastID != None:
    for p in currentSet:
        customers[p] = customers.get(p, 0) + 1
print(customers)
print(revenues)
```
